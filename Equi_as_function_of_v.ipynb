We work with a uniform distribution of $v$ between `v_l` and `v_h`. We choose values for $\theta_P,t_P,\theta_I,t_I$.

Below we choose quadratic functions for $c(v),\gamma(e)$

```python
v_l = 0.0
v_h =1.0
#theta_P = 0.9
t_P = 0.2
#theta_I = 0.5
t_I = 1.0


def F(v):
    return (v-v_l)/(v_h-v_l)

def f(v):
    return 1.0/(v_h-v_l)


C = 3.0
Gamma = 3.5

def c(v):
    return 0.5*C*v**2

def c_p(v):      # _p stands for "prime": derivative
    return C*v

def gamma(e):
    return 0.5*Gamma*e**2

def gamma_p(e):
    return Gamma*e

def gamma_pp(e):
    return Gamma


```

The rent function is given by 

$$
R_i = \int_0^{v_i} (p_i-c(v))f(v)dv = \int_0^{v_i} c'(v)F(v)dv
$$

$$
\frac{\partial \Pi_a}{\partial v_{a}} = f(v_a) \left[ (e+v_a-c(v_a)) (\frac{\theta_I}{3} + \frac{1-\theta_I}{2}) + \frac{\theta_I}{6}  c'(v_a) \frac{F(v_a)}{f(v_a)}  \right]
$$

$$
\frac{\partial \Pi_a}{\partial e} = \frac{1}{2} (1-\theta_I)F(v_{a}) - 2 \gamma'(e)
$$

```python
def R(v):
    return integrate.quad(lambda x: c_p(x)*F(x),0,v)[0]

def Pi_a_p_v(v,e,theta_I):
    return f(v)*((e+v-c(v))*(theta_I/3.0+(1-theta_I)/2.0)+theta_I/6.0*c_p(v)*F(v)/f(v))

def Pi_a_p_e(v,e,theta_I):
    return 0.5*(1-theta_I)*F(v)-2*gamma_p(e)
```

To find the equilibrium, we use the following equation from the paper:

\begin{equation}
R(v)F(v) \frac{\theta_P}{2\gamma t_P}+v-c(v) = \frac{F(v)}{f(v)} \frac{-\frac{\theta_I}{6}c'(v) + \frac{R(v)\frac{\theta_P}{t_P}-\frac{1}{2}
    (1-\theta_I)}{\gamma} \frac{\theta_P}{4 t_P} \frac{d(R(v)F(v))}{dv}}{\frac{\theta_I}{3}+\frac{1-\theta_I}{2}}
\end{equation}

Note that the assumption here is that $\gamma(e)=\frac{1}{2} \gamma e^2$


```python
def efficiency(v,theta_P,t_P,theta_I):
    return R(v)*F(v)*theta_P/(2.0*t_P*Gamma)+v-c(v)

def investment_effect(v,theta_P,t_P,theta_I):
    return F(v)/f(v)*(-theta_I*c_p(v)/6.0+(R(v)*theta_P/t_P-0.5*(1-theta_I))*(c_p(v)*F(v)**2+R(v)*f(v))*theta_P/(Gamma*4*t_P))/(theta_I/3.0+(1-theta_I)/2.0)
```

# Welfare

We plot $W$ as a function of $\theta_I,\theta_P$

```python
v_l = 0.0
v_h =1.0
#theta_P = 0.9
t_P = 0.2
#theta_I = 0.5
t_I = 1.0


def F(v):
    return (v-v_l)/(v_h-v_l)

def f(v):
    return 1.0/(v_h-v_l)


C = 3.0
Gamma = 3.5

def c(v):
    return 0.5*C*v**2

def c_p(v):      # _p stands for "prime": derivative
    return C*v

def gamma(e):
    return 0.5*Gamma*e**2

def gamma_p(e):
    return Gamma*e

def gamma_pp(e):
    return Gamma

def R(v):
    return integrate.quad(lambda x: c_p(x)*F(x),0,v)[0]

def Pi_a_p_v(v,e,theta_I):
    return f(v)*((e+v-c(v))*(theta_I/3.0+(1-theta_I)/2.0)+theta_I/6.0*c_p(v)*F(v)/f(v))

def Pi_a_p_e(v,e,theta_I):
    return 0.5*(1-theta_I)*F(v)-2*gamma_p(e)


def efficiency(v,theta_P,t_P,theta_I):
    return R(v)*F(v)*theta_P/(2.0*t_P*Gamma)+v-c(v)

def investment_effect(v,theta_P,t_P,theta_I):
    return F(v)/f(v)*(-theta_I*c_p(v)/6.0+(R(v)*theta_P/t_P-0.5*(1-theta_I))*(c_p(v)*F(v)**2+R(v)*f(v))*theta_P/(Gamma*4*t_P))/(theta_I/3.0+(1-theta_I)/2.0)

def FB(x): # x = [v,e]
    return [x[0]+x[1]-c(x[0]),F(x[0])-2*gamma_p(x[1])]

first_best = optimize.fsolve(FB,[1.5,1.5])

def welfare(v,e):
    return integrate.quad(lambda x: (x+e-c(x))*f(x),0,v)[0]-2*gamma(e)

def W(theta_P,theta_I):
    equil_v = optimize.brentq(lambda v: efficiency(v,theta_P,t_P,theta_I)+0.0001 - investment_effect(v,theta_P,t_P,theta_I), v_l,v_h)
    equil_e = R(equil_v)*F(equil_v)*theta_P/(2.0*t_P*Gamma)
    return welfare(equil_v,equil_e), equil_v, equil_e
    
print(first_best)
print(welfare(first_best[0],first_best[1]))


step = 0.01
range_v = arange(v_l,v_h+step,step)

def equilibrium_e(theta_P,theta_I):
    range_foc = [efficiency(v,theta_P,t_P,theta_I) - investment_effect(v,theta_P,t_P,theta_I) for v in range_v]
    if range_foc[1] < 0:
        print("lhs<rhs for v small, theta_P, theta_I="+str(theta_P)+", "+str(theta_I))
    elif any(x <0 for x in range_foc):
        index_negative = [n for n,i in enumerate(range_foc) if i<0.0 ][0]
        equil_v = optimize.brentq(lambda v: efficiency(v,theta_P,t_P,theta_I) - investment_effect(v,theta_P,t_P,theta_I),range_v[index_negative-1] ,range_v[index_negative]) 
    else:
        equil_v = 1.0
    equil_e = R(equil_v)*F(equil_v)*theta_P/(2.0*t_P*Gamma)
    return welfare(equil_v,equil_e), equil_v, equil_e



```
